# Show MD Simulation Movie

```python
import MDAnalysis as mda
import nglview as nv
import matplotlib.pyplot as plt

# 1. Load the universe
u = mda.Universe("md_0_10_ref.gro", "md_downsampled.xtc")

# Take every 10th frame for visualization
# u.trajectory[::10]

# 2. Setup the View
view = nv.show_mdanalysis(u)
view.clear()
view.parameters = {"backgroundColor": "white"}

# 3. PROTEIN: MANUAL RAINBOW (Index-Safe Slicing)
# We use Python to slice the protein into 20 color steps
n_segments = 20
cmap = plt.get_cmap('jet') # viridis, plasma
prot = u.select_atoms("protein")
resids = prot.residues.resids
step = len(resids) // n_segments

for i in range(n_segments):
    start_idx = i * step
    end_idx = (i + 1) * step if i < n_segments - 1 else len(resids)
    
    # Selection string using the residue numbers from the .gro file
    sel_str = f"protein and {resids[start_idx]}-{resids[end_idx-1]}"
    col = '#%02x%02x%02x' % tuple(int(c*255) for c in cmap(i/n_segments)[:3])
    
    view.add_representation("cartoon", selection=sel_str, color=col)

# 4. WATER: Faint Lines (Strictly restricted to 'O' or 'H' atoms)
# By adding 'element O or element H', we prevent it from grabbing the ions
view.add_representation("line", 
                        selection="(resname SOL or resname WAT) and (element O or element H)", 
                        # selection="(resname SOL or resname WAT) and (element O or element H) and (within 3.5 of protein)", 
                        color="lightblue", 
                        opacity=0.2)

# 5. IONS: The "Underscore" Fix (Guaranteed Purple)
# NGL uses '_Na' or '_NA' as a direct element pointer. 
# This is the most unambiguous way to select the ions.
view.add_representation("spacefill", 
                        selection="_Na or _NA", 
                        color="purple", 
                        radius=2.2)

# 6. Final Alignment
# view.add_unitcell()
view.center("protein")

view
```
